```python
%sql
use catalog demo;
use schema bronze;

CREATE OR REPLACE TABLE customer_reviews(
customer_id int, first_name string, last_name string, review string);

insert into customer_reviews values 
(1, 'John', 'Doe', 'Great product'),
(2, 'Jane', 'Doe', 'Good product'),
(3, 'Bob', 'Smith', 'Bad product'),
(4, 'Alice', 'Johnson', 'Excellent product'),
(5, 'Charlie', 'Brown', 'Okay product');

select * from customer_reviews;
```

```python
%sql
create function if not exists demo.bronze.full_name_udf(first_name string, last_name string)
returns string
return first_name || ' ' || last_name
```

```python
%sql
select customer_id, 
demo.bronze.full_name_udf(first_name, last_name) as full_name,
review
from demo.bronze.customer_reviews
```

```python
%sql
CREATE FUNCTION demo.bronze.sentiment_analysis(review string)
RETURNS string
RETURN 
  CASE
    WHEN review LIKE '%Great%' OR review LIKE '%Good%' OR review LIKE '%Excellent%' THEN 'Positive'
    WHEN review LIKE '%Bad%' THEN 'Negative'
    WHEN review LIKE '%Okay%' THEN 'Neutral'
    ELSE 'Unknown'
  END;
```

```python
%sql
SELECT customer_id,
  review,
  demo.bronze.sentiment_analysis(review) AS sentiment
FROM demo.bronze.customer_reviews
```
